Customizing Bar Graphs
Adding Titles, Legends and Resizing

Adding a title can be done by passing a title argument to the graph function. A legend can be added by setting the legend argument to true. We can also resize by passing a width and a height to the figsize argument. For instance, using the vehicles dataset of previous lessons:

```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline

vehicles = pd.read_csv('../data/vehicles.csv')
vehicles.head()
```

```python
vehicles['Drivetrain'].value_counts().plot(kind="bar", title="Count of Vehicles by Type of Drivetrain", 
                                           figsize=(15, 7), legend=True)
```

# Colors

When plotting a bar graph, we can use the standard color scheme but we can also use another color scheme.

Matplotlib has a long list of preset color maps.
The main issue is that these color maps have a set length that is sometimes longer than our data. In order to make sure that we get a diverse sampling from the color map, we must get the correct number of colors from the color map.

We do this by first finding out the number of unique values in our column:

```python
vehicles.Drivetrain.nunique()
```

Then we generate an array of values 0 through one minus the number of unique values in the column

```python
np.arange(vehicles.Drivetrain.nunique())
```

```python
plt.cm.Paired(np.arange(vehicles.Drivetrain.nunique()))
```

Finally, we plot our bar graph:

```python
vehicles['Drivetrain'].value_counts().plot(kind="bar", 
                                           color=plt.cm.Dark2(np.arange(vehicles.Drivetrain.nunique())),
                                           figsize=(15,7))
```

# Customizing Scatter Plots

Changing the Alpha

The alpha channel measures opaqueness. Alpha varies in the interval [0, 1], where the 0 value corresponds to full transparency, and 1 to opaqueness. Sometimes points tend to get clumped together. Reducing the alpha (i.e. increasing transparency) helps us to clearly detect distinct points even when there is a dense cluster of points.

Notice that in the plot below it is hard to tell how many points overlap. However, in the following plot with alpha=0.2, it is easier to tell.

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Fuel Barrels/Year", figsize=(15,7))
```

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Fuel Barrels/Year", 
                      alpha=0.2, figsize=(15,7))
```

# Adding a Grid

Having a grid can help read a scatter plot. In order to add a grid, we set the grid argument to true.

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Fuel Barrels/Year", 
                      alpha=0.2, grid=True, figsize=(15,7))
```
