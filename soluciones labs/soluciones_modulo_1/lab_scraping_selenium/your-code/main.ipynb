####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code

driver=webdriver.Chrome(PATH)

driver.get(url)

tabla=driver.find_element_by_xpath('//*[@id="tbody"]')


date=[e.text.split('\n')[0].split()[0] for e in tabla.find_elements_by_class_name('tabev6')]

tiempo=[e.text.split('\n')[0].split()[1] for e in tabla.find_elements_by_class_name('tabev6')]

lat_lng=[e.text for e in tabla.find_elements_by_class_name('tabev1')]
lat_lng_coor=[e.text.strip() for e in tabla.find_elements_by_class_name('tabev2') if len(e.text.strip())==1]

lat_lng=[''.join(e) for e in zip(lat_lng, lat_lng_coor)]

lat=[lat_lng[i] for i in range(len(lat_lng)) if i%2==0]
lng=[lat_lng[i] for i in range(len(lat_lng)) if i%2==1]


region_name=[e.text.strip() for e in tabla.find_elements_by_class_name('tb_region')]

driver.quit()

columnas=['date', 'time', 'latitude', 'longitude' , 'region_name']
data=[date, tiempo, lat, lng, region_name]

df=pd.DataFrame(columns=columnas)

for i in range(len(data)):
    df[columnas[i]]=data[i]
    
df
````

#### Book name,price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
#your code

driver=webdriver.Chrome(PATH)

driver.get(url)

time.sleep(3)

tabla=driver.find_element_by_xpath('//*[@id="default"]/div[2]/div/div/div/section/div[2]/ol')

libros=tabla.find_elements_by_tag_name('li')

name=[e.text.split('\n')[0] for e in libros]

price=[e.text.split('\n')[1] for e in libros]

stock=[e.text.split('\n')[2] for e in libros]

driver.quit()

df=pd.DataFrame(columns=['name', 'price', 'stock'])

df.name=name

df.price=price

df.stock=stock

df
````
