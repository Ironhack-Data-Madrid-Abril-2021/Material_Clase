**Primero cargar datos...**

```python
import pandas as pd

df=pd.read_csv('../data/shoes.csv')

df.head()
```

```python
df=df[['brand', 'colors', 'sizes', 'prices.amountMax', 'imageURLs', 'ean', 'upc']]  # seleccionar columnas

df.head()
```

```python
df=df.dropna().reset_index(drop=True)   # quitar nulos

df.head()
```

```python
df.info()
```

```python
df.sizes=df.sizes.apply(lambda x: x.split(','))
df.colors=df.colors.apply(lambda x: x.split(','))

df.head()
```

```python
df=df.explode('sizes').reset_index(drop=True)
df=df.explode('colors').reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
codigo=pd.read_csv('../data/upc_corpus.csv')
codigo=codigo.dropna().reset_index(drop=True)

codigo.head()
```

```python
codigo.info()
```

```python
df.ean=codigo.ean.iloc[0:len(df)]

df.head()
```

```python
df=df.rename(columns={'colors':'color', 'sizes':'size', 'prices.amountMax': 'prices', 'imageURLs':'url'})

df.head()
```

```python
df['id']=[i for i in range(len(df))]

df.head()
```

```python
import mysql.connector as conn
```

```python
# conexion al servidor de sql, para crear base de datos

crea_db=conn.connect(host='localhost', user='root', passwd='password')

cursor=crea_db.cursor()

cursor.execute('drop database if exists productos')  # si existe la base de datos, borrala

cursor.execute('create database productos')   # crea la base de datos
```

```python
# check

cursor.execute('show databases')
for x in cursor:
    print(x)
```

```python
# conectar a sql, a una db especifica

db=conn.connect(host='localhost', user='root', passwd='password', database='productos')

cursor=db.cursor()
```

```python
# crear una tabla

cursor.execute('drop table if exists zapatos')   # borra la tabla si existe



tabla='''
        create table zapatos(
            
            id int,
            brand varchar(50),
            color varchar(50),
            size varchar(50),
            prices float,
            url varchar(10000),
            ean varchar(50),
            upc varchar(5000)
        );
'''


cursor.execute(tabla)
```

```python
# check

cursor.execute('show tables')
for x in cursor:
    print(x)
```
